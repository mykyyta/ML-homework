```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#%config InlineBackend.figure_format = 'svg'
```

```python
players = pd.read_csv('/content/drive/MyDrive/Machine Learning/data/hw_vis/Players.csv', index_col=0)
seasons = pd.read_csv('/content/drive/MyDrive/Machine Learning/data/hw_vis/Seasons_Stats.csv', index_col=0)
player_data = pd.read_csv('/content/drive/MyDrive/Machine Learning/data/hw_vis/player_data.csv')
```

```python
#visuzalising missing data (1.)
plt.figure(figsize = (10, 20))
plt.tick_params(axis='both', which='major', labelsize=8, labelbottom = True, bottom=False, top = False, labeltop=True)
sns.heatmap(seasons.isna(), cbar=False)
```

```python
#removing blank rows and columns
seasons.dropna(how='all', axis=0, inplace=True)
seasons.drop(seasons[['blanl', 'blank2']], inplace=True, axis=1)
```

```python
#separate numeric columns
seasons_num = seasons.select_dtypes(include=[np.number])
```

```python
seasons_num.shape
```

```python
seasons_num.head()
```

```python
#boxplots of all numeric values, overview (2.)
seasons_num.plot.box(subplots=True, layout=(6, 8), figsize=(16, 17),
                     sharex=False, sharey=False, legend=False, title=list(seasons_num.columns),
                     flierprops={'marker': 'o', 'markersize': 3})
plt.tight_layout()
```

```python
#(3)
fig, axes = plt.subplots(8, 6, figsize=(15, 15), constrained_layout=True)

for col, ax in zip(seasons_num.columns, axes.flat):
    seasons_num[col].plot.hist(ax=ax, rot=0, bins=20)
    ax.set_title(col)

for ax in axes.flat[seasons_num.columns.size:]:
    ax.set_axis_off()
```

```python
# creating correlation matrix
seasons_num_sample = seasons_num.sample(frac=0.1, random_state=100)
corr=seasons_num_sample.corr()
matrix = np.triu(corr)
corr = corr.round(1)
```

```python
# plotting correlation matrix (4.)
plt.figure(figsize = (25, 13))
plt.tick_params(axis='both', which='major', labelsize=10, labelbottom = True, bottom=False, top = False, labeltop=True, labelright=True)
sns.heatmap(corr, mask=matrix, annot=True)
```

```python
# relationship between selected variables (5.)
sns.set_style('darkgrid')
sns.pairplot(data=seasons_num_sample, vars=['OWS', 'DWS', 'WS', 'WS/48', 'TOV%', 'TOV'], kind='kde')
```

```python
# points scored by teams. (6.) 'TOT' team has highest score with huge disctance to second team.
#After small investigation I found out that 'TOT' means 'total' and it is sum of several rows
#when player played in diffirent teams during one season. So if we want to avoid repetitions in dataset we have to drop rows with 'TOT' team
fig, ax = plt.subplots(figsize = (13, 7))
team_pts = seasons.groupby('Tm')['PTS'].sum().sort_values(ascending=False)
sns.barplot(team_pts)
ax.tick_params(axis='x', labelrotation=90)
```

```python
# dropping TOT
seasons_clean = seasons.drop(seasons[seasons['Tm'] == 'TOT'].index)
seasons_clean.head(20)
```

```python
# sum of points, distributuion by age (7.)
plt.figure(figsize = (13, 7))
sns.barplot(data=seasons_clean, x='Age', y='PTS', estimator="sum", errorbar=None)
```

```python
# distribution of career points by all players (8.)
import matplotlib.ticker as mtick


all_players = seasons_clean.groupby('Player')['PTS'].sum().sort_values(ascending=False)
all_players = pd.Series(all_players)

fig, ax = plt.subplots(figsize = (13, 7))
all_players.plot()
ax.set_xticks(np.linspace(1, len(all_players), num=21))
ax.xaxis.set_major_formatter(mtick.PercentFormatter(len(all_players)))
```

```python
# top20 players by points (9.)
top_20 = seasons_clean.groupby('Player')['PTS'].sum().sort_values(ascending=False)[:20]
fig, ax = plt.subplots()
sns.barplot(top_20)
ax.set_ylim([25000, 40000])
ax.tick_params(axis='x', labelrotation=90)
```

```python
# filtering rows with data of top_20 players
top_20_names = list(top_20.index)
seasons_top_20 = seasons[seasons['Player'].isin(top_20_names)].select_dtypes(include=[np.number])
```

```python
# boxplot  + mean value of these feature for top_20 players (10.)
fig, axes = plt.subplots(5, 10, figsize=(15, 20), constrained_layout=True)

for col_1, col_2, ax in zip(seasons_num.columns, seasons_top_20.columns, axes.flat):
    sns.boxplot(seasons_num[col_1], ax=ax)
    ax.axhline(y=seasons_top_20[col_2].mean(), color='purple')
    ax.set_title(col_1)

for ax in axes.flat[seasons_num.columns.size:]:
    ax.set_axis_off()
```
